```python
# Importing required packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Read training Mercedes Benz data
benz_train_df = pd.read_csv("train.csv")
benz_test = pd.read_csv("test.csv")
```

```python
benz_train_df.head()
```

```python
benz_test.head()
```

```python
# Check number of columns and rows
print("Shape of training data", benz_train_df.shape)
print("Shape of test data", benz_test.shape)
```

```python
# Check data types
benz_train_df.dtypes
```

```python
# Removing ID column as it has no relation to target
benz_train_df.drop(['ID'], axis=1, inplace=True)
benz_test.drop(['ID'], axis=1, inplace=True)
```

```python
# Removing duplicates
benz_train_df.duplicated().sum()
```

```python
benz_train_df.drop_duplicates(inplace=True)
```

```python
benz_train_df.shape
```

```python
# Identifying column 'y' as target and dropping from train ds
target = benz_train_df['y']
print(target.describe(), "\n Shape: ", target.shape)
```

```python
# Removing target column from train ds
benz_train_df.drop(['y'], axis=1, inplace=True)
benz_train_df.head(2)
```

```python
for col in benz_train_df.columns:
    
    # Checking the unique values in Training and test data set
    train_unique =benz_train_df[col].nunique()
    test_unique = benz_test[col].nunique()
    print("Train dataset - Number: {} \t & Unique values in Column {} :: ".format(train_unique, col), benz_train_df[col].unique())
    print("Test dataset - Number: {} \t & Unique values in Column {} :: ".format(test_unique, col), benz_test[col].unique())
    
    # Removing test data values which are not present in training data
    if (benz_test[col].dtype == 'object'):
        for val in benz_test[col].values:
            if(val not in benz_train_df[col].values):
                benz_test.drop(benz_test[benz_test[col] == val].index, inplace=True)
    
```

# * Apply label encoder.

# * If for any column(s), the variance is equal to zero, then you need to remove those variable(s).

```python
# transform categorical columns into integer values 
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in benz_train_df.columns :
    if(benz_train_df[col].dtype == 'object') :
            benz_train_df[col] = le.fit_transform(benz_train_df[col])
            benz_test[col] = le.transform(benz_test[col])
            
# Find variance of each column
    if (benz_train_df[col].var() == 0):
        print("Dropping column: {}".format(col))
        # If there is only 1 unique value in a column - that means there is no variance - dropping these columns
        benz_train_df.drop(col, axis=1, inplace=True)
        benz_test.drop(col, axis=1, inplace=True)
```
